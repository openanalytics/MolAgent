### Loading two models

```python
##################################################
#the filename of the model
save_model_to_file='Caco2_Wang_stackingregmodel.pt'
##################################################
from automol.stacking import load_model

stacked_model1= load_model( save_model_to_file,use_gpu=False)
stacked_model2= load_model( save_model_to_file,use_gpu=False)
print(stacked_model1.reproducable_output.keys())
print(stacked_model2.reproducable_output.keys())
```

### Deleting properties

```python
properties=[p for p in stacked_model1.models]
stacked_model1.delete_properties(properties[0])
stacked_model2.delete_properties(properties[1])
stacked_model1.generate_reproducable_output()
stacked_model2.generate_reproducable_output()
print(stacked_model1.reproducable_output.keys())
print(stacked_model2.reproducable_output.keys())
```

### Merging models

```python
stacked_model1.merge_model(other_model=stacked_model2,other_props=[properties[0]])
stacked_model1.print_metrics()
print(stacked_model1.reproducable_output.keys())
print(stacked_model2.reproducable_output.keys())
```

### Cleaning and saving for deployment

```python
from jnj_auto_ml.stacking import save_model
stacked_model1.deep_clean()
save_model(stacked_model1, save_model_to_file,create_reproducability_output=True)
```
