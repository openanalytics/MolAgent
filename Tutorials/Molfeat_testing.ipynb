### molfeat

Molfeat is a package collecting al kinds of features for molecules. In this notebook we show how to use the automol wrappers for the different kind of features in molfeat. For details see [molfeat](https://molfeat.datamol.io/featurizers). Note that for all the 3D features shown here, the EmbedMolecule functionality of RDKIT is used. 

You can see how to add them to your training in the notebooks, in short the feature generators are stored in a dictionary in the stacking_model in the variable **feature_generators**. Just add your generator with a key and use that key in the selected features argument.

```python
from automol.feature_generators import MolfeatPretrainedHFTransformer
# [kind='ChemGPT-4.7M', notation='selfies'], [kind='ChemBERTa-77M-MTR', notation='smiles'], [kind='ChemBERTa-77M-MLM', notation='smiles'],
# [kind='MolT5', notation='smiles'] ,[kind='ChemGPT-1.2B', notation='selfies'], [kind='ChemBERTa-77M-MTR', notation='smiles'], [kind='ChemGPT-4.7M', notation='selfies']
feature_generator= MolfeatPretrainedHFTransformer( kind='ChemGPT-19M', notation='selfies',dtype=float,max_length=220)
X=feature_generator.generate(['[Cu]', 'C[C@H](N)C(=O)O', 'CC(O)C(=O)O', 'OCC(O)CO', 'Oc1ccccc1','Nc1ccncc1', None, 'C#CC(C)(O)CC', '', 'ClCCCl', 'O=C1CCC(=O)N1', 'O=CCCCC=O','N[C@@H]1CONC1=O', 'S1C=CSC1=C'])
X.shape
```

```python
from automol.feature_generators import MolfeatFPVecTransformer
# desc2D, atompair-count, topological-count, fcfp-count, ecfp-count, estate, erg, secfp, pattern, rdkit, fcfp, ecfp, avalon, maccs
feature_generator= MolfeatFPVecTransformer(kind='ecfp', dtype=float)
X=feature_generator.generate(['[Cu]', 'C[C@H](N)C(=O)O', 'CC(O)C(=O)O', 'OCC(O)CO', 'Oc1ccccc1','Nc1ccncc1', None, 'C#CC(C)(O)CC', '', 'ClCCCl', 'O=C1CCC(=O)N1', 'O=CCCCC=O','N[C@@H]1CONC1=O', 'S1C=CSC1=C'])
X.shape
```

```python
from automol.feature_generators import Molfeat3DFPVecTransformer
#fails: map4
# desc3D, electroshape, usrcat (without Cu), usr (without Cu)
feature_generator= Molfeat3DFPVecTransformer(kind='electroshape', dtype=float)
X=feature_generator.generate(['[Cu]', 'C[C@H](N)C(=O)O', 'CC(O)C(=O)O', 'OCC(O)CO', 'Oc1ccccc1','Nc1ccncc1', None, 'C#CC(C)(O)CC', '', 'ClCCCl', 'O=C1CCC(=O)N1', 'O=CCCCC=O','N[C@@H]1CONC1=O', 'S1C=CSC1=C'])
X.shape
```

```python
from automol.feature_generators import MolfeatMoleculeTransformer
from molfeat.calc.pharmacophore import Pharmacophore3D
from molfeat.calc.pharmacophore import Pharmacophore2D

# mordred, scaffoldkeys, cats2d, Pharmacophore2D(factory='cats'), Pharmacophore2D(factory='gobbi'), Pharmacophore2D(factory='pmapper'), 
feature_generator= MolfeatMoleculeTransformer(featurizer='cats2d', dtype=float)
X=feature_generator.generate(['[Cu]', 'C[C@H](N)C(=O)O', 'CC(O)C(=O)O', 'OCC(O)CO', 'Oc1ccccc1','Nc1ccncc1', None, 'C#CC(C)(O)CC', '', 'ClCCCl', 'O=C1CCC(=O)N1', 'O=CCCCC=O','N[C@@H]1CONC1=O', 'S1C=CSC1=C'])
X.shape
```

```python
from automol.feature_generators import Molfeat3DMoleculeTransformer
from molfeat.calc.pharmacophore import Pharmacophore3D

# Pharmacophore3D(factory='cats'), 'cats3d', Pharmacophore3D(factory='default'),Pharmacophore3D(factory='gobbi')
feature_generator= Molfeat3DMoleculeTransformer(featurizer=Pharmacophore3D(factory='pmapper'), dtype=float)
X=feature_generator.generate(['C1CC(=O)NC(=O)C1N2C(=O)C3=CC=CC=C3C2=O','[Cu]', 'C[C@H](N)C(=O)O', 'CC(O)C(=O)O', 'OCC(O)CO', 'Oc1ccccc1','Nc1ccncc1', None, 'C#CC(C)(O)CC', '', 'ClCCCl', 'O=C1CCC(=O)N1', 'O=CCCCC=O','N[C@@H]1CONC1=O', 'S1C=CSC1=C'])
X.shape
```
